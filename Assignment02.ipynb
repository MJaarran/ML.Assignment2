```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA

sns.set(style="whitegrid")

df = pd.read_csv('SuperMarket Analysis.csv')

print(df.info())

df['Date'] = pd.to_datetime(df['Date'])

df_missing = df.copy()
df_missing.loc[0:10, 'Sales'] = np.nan

median_sales = df_missing['Sales'].median()
df_missing['Sales'] = df_missing['Sales'].fillna(median_sales)

Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)].copy()

features_to_scale = ['Sales', 'Unit price']
scaling_df = df_cleaned[features_to_scale].copy()

min_max_scaler = MinMaxScaler()
df_min_max = pd.DataFrame(min_max_scaler.fit_transform(scaling_df), columns=features_to_scale)

std_scaler = StandardScaler()
df_zscore = pd.DataFrame(std_scaler.fit_transform(scaling_df), columns=features_to_scale)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_zscore)

print(pca.explained_variance_ratio_)
```
